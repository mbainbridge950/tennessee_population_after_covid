```python
county_city_tn.rename(columns={'County Seat': 'County', 'Other City':'City'}, inplace=True)


city_county_long = county_city_tn.melt(
    id_vars=['County'],                   
    value_vars=county_city_tn.columns[1:], 
    value_name='City'
)


city_county_long = city_county_long[['County','City']].dropna()

county_seat_df = county_city_tn[['CountySeat','County']].copy()

county_seat_df.rename(columns={'CountySeat':'City'}, inplace=True) 
city_county_long = pd.concat([city_county_long, county_seat_df[['County','City']]], ignore_index=True)

city_county_long['City'] = city_county_long['City'].str.strip()
city_county_long['County'] = city_county_long['County'].str.strip()


city_county_long = city_county_long.drop_duplicates()

# Check result
city_county_long.head(15)
```

```python
city_county_long.to_csv('../data/city_county_long.csv',index=False)
```
