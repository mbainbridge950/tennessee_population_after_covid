```python

covid_stats['WEEK_END_DATE'] = pd.to_datetime(covid_stats['WEEK_END_DATE'])

# Filter for Blount County
blount = covid_stats[covid_stats['COUNTY'] == 'Blount']

# Get last week of each year
blount_year_end = blount.sort_values('WEEK_END_DATE').groupby('Year').tail(1).reset_index(drop=True)

# Compute actual deaths per year
blount_year_end['YearlyDeaths'] = blount_year_end['TOTAL_DEATHS'].diff().fillna(blount_year_end['TOTAL_DEATHS'])

# Total deaths across all years
total_deaths_all_years = blount_year_end['YearlyDeaths'].sum()

print("Total deaths in Blount County (2020-2024):", int(total_deaths_all_years))
```

```python

covid_stats['WEEK_END_DATE'] = pd.to_datetime(covid_stats['WEEK_END_DATE'])

# Filter for Blount County
davidson = covid_stats[covid_stats['COUNTY'] == 'Davidson']

# Get last week of each year
davidson_year_end = davidson.sort_values('WEEK_END_DATE').groupby('Year').tail(1).reset_index(drop=True)

# Compute actual deaths per year
davidson_year_end['YearlyDeaths'] = davidson_year_end['TOTAL_DEATHS'].diff().fillna(davidson_year_end['TOTAL_DEATHS'])

# Total deaths across all years
total_deaths_all_years = davidson_year_end['YearlyDeaths'].sum()

print("Total deaths in Davidson County (2020-2024):", int(total_deaths_all_years))
```

```python
covid_stats['WEEK_END_DATE'] = pd.to_datetime(covid_stats['WEEK_END_DATE'])

# Get last week of each year per county
year_end = (
    covid_stats.sort_values(['COUNTY', 'Year', 'WEEK_END_DATE'])
    .groupby(['COUNTY', 'Year'])
    .tail(1)
    .reset_index(drop=True)
)

# Compute year-over-year deaths for each county
year_end['YearlyDeaths'] = year_end.groupby('COUNTY')['TOTAL_DEATHS'].diff().fillna(year_end['TOTAL_DEATHS'])

# Compute total deaths across all years per county
total_deaths_by_county = year_end.groupby('COUNTY')['YearlyDeaths'].sum().reset_index()


#Find the county with the highest deaths
highest_deaths_county = total_deaths_by_county.loc[total_deaths_by_county['YearlyDeaths'].idxmax()]

print("County with highest total deaths (2020-2024):")
print(highest_deaths_county)
```

```python
total_deaths_by_county = year_end.groupby('COUNTY')['YearlyDeaths'].sum().reset_index()
```

```python
total_deaths_state = total_deaths_by_county['YearlyDeaths'].sum()
```

```python
total_deaths_by_county['DeathPercentage'] = (total_deaths_by_county['YearlyDeaths'] / total_deaths_state) * 100
```

```python
highest_percentage_county = total_deaths_by_county.loc[total_deaths_by_county['DeathPercentage'].idxmax()]
print("County with highest percentage of COVID deaths:")
print(highest_percentage_county)
```
